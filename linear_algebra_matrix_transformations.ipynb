```python
# Let's visualize a simple matrix transformation

import numpy as np
import matplotlib.pyplot as plt

# Create a simple vector
v = np.array([2, 1])

# Create a transformation matrix (scaling and rotation)
A = np.array([[1.5, 0.5], 
              [-0.5, 1.5]])

# Apply transformation
v_transformed = A @ v  # Matrix multiplication

# Plot both vectors
fig, ax = plt.subplots()
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Original vector')
ax.quiver(0, 0, v_transformed[0], v_transformed[1], angles='xy', scale_units='xy', scale=1, color='red', label='Transformed vector')
ax.set_xlim(-3, 5)
ax.set_ylim(-3, 5)
ax.axhline(y=0, color='k', linewidth=0.5)
ax.axvline(x=0, color='k', linewidth=0.5)
ax.grid(True, linestyle='--', alpha=0.7)
ax.legend()
ax.set_aspect('equal')
plt.title("Matrix Transformation Visualization")
plt.show()

print(f"Original vector: {v}")
print(f"Transformed vector: {v_transformed}")
print("\nThis is how I teach MAT 322: Linear Algebra - showing the geometry behind the algebra.")
```
